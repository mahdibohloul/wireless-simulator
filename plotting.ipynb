# Plotting the data extracted from the simulation

### Defining the figure

To create a graph, we need to fix the bandwidth and construct graphs with the error_rate bars.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('analyzer_result.csv')
df
```

## Throughput Analysis

### Throughput with bandwidth 1.5MB

```python
df_1_5 = df[df['bandwidth'] == 1.5]
df_1_5.plot(x='error_rate', y='throughput')
plt.xlabel('Error Rate')
plt.ylabel('Throughput')
plt.title('Throughput with bandwidth 1.5MB')
plt.show()
```

### Throughput with bandwidth 55MB

```python
df_55 = df[df['bandwidth'] == 55]
df_55.plot(x='error_rate', y='throughput')
plt.xlabel('Error Rate')
plt.ylabel('Throughput')
plt.title('Throughput with bandwidth 55MB')
plt.show()
```

### Throughput with bandwidth 155MB

```python
df_155 = df[df['bandwidth'] == 155]
df_155.plot(x='error_rate', y='throughput')
plt.xlabel('Error Rate')
plt.ylabel('Throughput')
plt.title('Throughput with bandwidth 155MB')
plt.show()
```
